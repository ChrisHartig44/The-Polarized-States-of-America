```
import numpy as np

conditions = [
    (df['Margin'] >= 10),
    (df['Margin'] >= 5) & (df['Margin'] < 10),
    (df['Margin'] > -5) & (df['Margin'] < 5),
    (df['Margin'] > -10) & (df['Margin'] <= -5),
    (df['Margin'] <= -10),
    ]
values = ['Safe Democrat', 'Lean Democrat', 'Swing', 'Lean Republican', 'Safe Republican',]
df['Status'] = np.select(conditions, values)
```

```
color_dict = dict({'Safe Democrat':'blue',
                  'Lean Democrat':'slateblue',
                  'Swing Democrat': 'purple',
                  'Lean Republican': 'indianred',
                   'Safe Republican': 'red'})
```

```
import plotly.express as px
fig = px.scatter(df, x= 'Margin', y= 'Change In Partisan Lean', color="Status",
                 size='Electoral Votes', hover_name="State",color_discrete_map={'Safe Democrat':'blue',
                  'Lean Democrat':'slateblue',
                  'Swing': 'purple',
                  'Lean Republican': 'indianred',
                   'Safe Republican': 'red'}, title="The Polarized States of America", size_max=80, animation_group='State', animation_frame='Year', range_x= [-90,90], range_y=[-40,40]

)
fig.update_layout(
    margin=dict(l=11, r=11, t=11, b=11),
    paper_bgcolor="white",
)
fig.show()
```
